```python
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
template = "Write down the following linear programming problem in mathematical form:"

text = "Let the company have a budget of $ 10,000, there are also 4 areas in which you can invest money. The first area is employee salaries, the second is business expansion, the third is advertising, and the fourth is hiring new employees. This month, it is necessary to financethe first and second spheres as much as possible. At the same time, there are some additional conditions: the difference between the firstand second spheres should be >= 100$, between the second and third >= 200$, and between the third andfourth >= 300$."
sequences = pipeline(
    template + text
)
```

```python
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
